```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

# Initialize vectors
taus = list()
time = np.arange(0, 1, 0.0001)

# Load data
spks = np.loadtxt('spikes.dat', dtype='int16')
stim = np.loadtxt('stimulus.dat', dtype=float)[:-1,1]

# Find ISI from spike trains
for spk in spks:
    aux = time[(spk!=0)]
    taus.append([aux[i+1]-aux[i] for i in range(len(aux)-1)])

# Count spikes in trials
ns = [len(tau) for tau in taus]

# Plot ISI probability density
plt.hist([inner for outer in taus for inner in outer], 200, density=True, histtype='step', rwidth=0.85, color='b')
plt.title(r'Probability Distribution for ISI', color='blue', fontsize=34)
plt.xlabel(r'$\tau$ [ms]')
plt.ylabel(r'$p(\tau)$')
plt.show()

# Mean and standard deviation
stds = [ np.std(tau) for tau in taus]
means = [ np.mean(tau) for tau in taus]

# Variavility coefficient
cvs = [std/mean for std, mean in zip(stds, means)]
plt.plot(range(len(cvs)), cvs, 'bo')
plt.title(r'Coeff. de variabilidad CV', color='blue', fontsize=34)
plt.xlabel(r'$Trials$ [ms]')
plt.ylabel(r'$CV$')
plt.show()
```

```python
# Fano factor
ff = np.var(ns)/np.mean(ns)


plt.hist(ns, range(85,145), density=True, histtype='step', rwidth=0.85, color='b', label="Fano: "+str(ff))
plt.title(r'Probability Distribution for N', color='blue', fontsize=34)
plt.xlabel(r'N')
plt.ylabel(r'$p(N)$')
plt.legend()
plt.show()
```

```python
plt.plot(time[1:], np.sum(spks, axis=0)[1:]/time[1]/len(spks), linewidth=0.75)
plt.title(r'Firing rate vs time', color='blue', fontsize=34)
plt.xlabel(r'time [sec]')
plt.ylabel(r'Frequency [Hz]')
plt.show()
```

```python
# spike time
indices = [ti for spk in spks for ti in np.arange(10000)[(spk!=0)]]

# initialize filter
Qrs = np.zeros(len(stim))

stim -= np.mean(stim)

# calculate Qrs
for index in indices: Qrs[:index+1] += stim[index::-1] 

plt.plot(time[:500], Qrs[:500], '-o')
plt.title(r'Filter', color='blue', fontsize=34)
plt.xlabel(r'time [sec]')
plt.ylabel(r'UA')
plt.show()
```
